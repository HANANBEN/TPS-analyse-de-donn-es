```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
data1 = pd.read_csv("Automobile_data.csv")
data1=data1[['highway-mpg', 'engine-size', 'horsepower', 'curb-weight','price']]
for x,row in data1.iterrows():
    if row['engine-size']=='?':
        data1.drop(x,inplace=True) 
for x,row in data1.iterrows():
    if row['horsepower']=='?':
        data1.drop(x,inplace=True)
for x,row in data1.iterrows():
    if row['highway-mpg']=='?':
        data1.drop(x,inplace=True)
for x,row in data1.iterrows():
    if row['curb-weight']=='?':
        data1.drop(x,inplace=True)
for x,row in data1.iterrows():
    if row['price']=='?':
        data1.drop(x,inplace=True)      
X = data1[['highway-mpg', 'engine-size', 'horsepower', 'curb-weight']]
y = data1['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("Score R²: {:.2f}".format(r2))
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], '--', color='red')
plt.xlabel('Valeurs réelles')
plt.ylabel('Prédictions')
plt.show()
```
