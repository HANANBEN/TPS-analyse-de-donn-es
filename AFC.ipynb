```python
import pandas as pd
D=pd.read_excel("data.xlsx",index_col=0)
D
```

```python
from sklearn.decomposition import FactorAnalysis
afc = FactorAnalysis()
resultats_afc = afc.fit_transform(D)
valeurs_propres = afc.noise_variance_
vecteurs_propres = afc.components_
print("Valeurs propres:", valeurs_propres)
print("Vecteurs propres:", vecteurs_propres)
```
